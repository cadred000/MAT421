## Numerical Differentiation Problem Statement

In Python a numerical grid is created with evenly spaced points over a given domain.
Then, a function is defined and its value is calculated at each point in the grid.
The problem is how to approximate the derivative of the function at a given point.

## Finite Difference Approximating Derivatives

We will use the forward difference formula.

```python
import numpy as np
import matplotlib.pyplot as plt

h = 0.01
x = np.arange(0, 2*np.pi, h)
f = np.cos(2*x)
df = -2*np.sin(2*x)

f_diff = np.diff(f)/h
f_diff = np.append(f_diff, f_diff[-1])
x_diff = x[:-1:]

plt.figure(figsize=(10, 6))
plt.plot(x, f_diff, '--', label='Approximate', color='blue')
plt.plot(x, df, label='Exact', linestyle='-', color='red')
plt.title('Comparison of Derivative Approximations')
plt.xlabel('$x$')
plt.ylabel('Derivative')
plt.legend()
plt.grid(True)
plt.show()
```

## Approximating Higher Order Derivatives

Taylor series approximation using Sympy

```python
import sympy as sp

x = sp.symbols('x')
f = sp.cos(2*x)

n = 4
taylor = f.series(x, 1, n).removeO()

num_taylor = sp.lambdify(x, taylor, 'numpy')

domain = np.linspace(0,np.pi, 250)
taylor_dom = np.cos(2*domain)
num_dom = num_taylor(domain)

plt.figure(figsize=(10, 6))
plt.plot(domain, taylor_dom, label='Exact $\\cos(2x)$', linestyle='-', color='blue')
plt.plot(domain, num_dom, label=f'Taylor Series Approximation (n={n-1})', linestyle='--', color='red')
plt.title('Comparison of $\\cos(2x)$ and Its Taylor Series Approximation')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.grid(True)
plt.show()
```
